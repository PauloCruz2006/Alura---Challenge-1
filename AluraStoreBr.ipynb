# 2. Vendas por Categoria


##2.1. Atividade

Neste passo, deve calcular a quantidade de produtos vendidos por categoria em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

```python
import pandas as pd

# URLs dos arquivos CSV no GitHub
url_loja1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv"
url_loja2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv"
url_loja3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv"
url_loja4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"

# Carregando os dados
loja1 = pd.read_csv(url_loja1)
loja2 = pd.read_csv(url_loja2)
loja3 = pd.read_csv(url_loja3)
loja4 = pd.read_csv(url_loja4)

# Função para contar vendas por categoria
def vendas_por_categoria(df, nome_loja):
    resumo = df.groupby("Categoria do Produto")["Produto"].count().reset_index()
    resumo = resumo.rename(columns={"Produto": "Quantidade Vendida"})
    resumo = resumo.sort_values(by="Quantidade Vendida", ascending=False)
    print(f"\n=== {nome_loja} ===")
    print(resumo)
    return resumo

# Aplicando para cada loja
vendas_loja1 = vendas_por_categoria(loja1, "Loja 1")
vendas_loja2 = vendas_por_categoria(loja2, "Loja 2")
vendas_loja3 = vendas_por_categoria(loja3, "Loja 3")
vendas_loja4 = vendas_por_categoria(loja4, "Loja 4")
```

# 3. Média de Avaliação das Lojas

##3.1. Atividade

Neste passo, vamos calcular a média das avaliações dos clientes para cada loja. O objetivo é entender a satisfação dos clientes com os produtos vendidos.

```python
import pandas as pd

# URLs dos arquivos CSV
url_loja1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv"
url_loja2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv"
url_loja3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv"
url_loja4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"

# Carregando os dados
loja1 = pd.read_csv(url_loja1)
loja2 = pd.read_csv(url_loja2)
loja3 = pd.read_csv(url_loja3)
loja4 = pd.read_csv(url_loja4)

# Calculando a média das avaliações (usando o nome correto da coluna)
coluna_avaliacao = "Avaliação da compra"

media_loja1 = loja1[coluna_avaliacao].mean()
media_loja2 = loja2[coluna_avaliacao].mean()
media_loja3 = loja3[coluna_avaliacao].mean()
media_loja4 = loja4[coluna_avaliacao].mean()

# Exibindo resultados
print(f"Média de avaliação - Loja 1: {media_loja1:.2f}")
print(f"Média de avaliação - Loja 2: {media_loja2:.2f}")
print(f"Média de avaliação - Loja 3: {media_loja3:.2f}")
print(f"Média de avaliação - Loja 4: {media_loja4:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos arquivos CSV
urls = {
    "Loja 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv",
    "Loja 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv",
    "Loja 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv",
    "Loja 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"
}

coluna_avaliacao = "Avaliação da compra"
medias = {loja: pd.read_csv(url)[coluna_avaliacao].mean() for loja, url in urls.items()}

# Parâmetros
meta_min, meta_ideal, limite = 4.0, 4.5, 5

# Criar gráfico
fig, ax = plt.subplots(figsize=(8, 4))

# Índices para manter ordem Loja 1 -> Loja 4
y_pos = list(range(len(medias)))
for i, loja in enumerate(medias.keys()):
    media = medias[loja]
    # fundo cinza
    ax.barh(i, limite, color="#e0e0e0", height=0.4)
    # barra azul real
    ax.barh(i, media, color="#4E79A7", height=0.4)
    # linhas de meta
    ax.vlines(meta_min, i - 0.2, i + 0.2, color="orange", linestyle="--", linewidth=2)
    ax.vlines(meta_ideal, i - 0.2, i + 0.2, color="green", linestyle="-", linewidth=2)
    # rótulo de valor
    ax.text(media + 0.05, i, f"{media:.2f}", va='center', fontsize=9, fontweight='bold')

# Ajustes visuais
ax.set_yticks(y_pos)
ax.set_yticklabels(medias.keys())
ax.set_xlim(0, limite)
ax.set_xlabel("Média de Avaliação (0 a 5)")
ax.set_title("Desempenho das Lojas - Avaliação Média", fontsize=14, fontweight='bold')
ax.grid(axis="x", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

##4.1. Atividade

Neste passo, deve identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja.  Ao final, visualize os resultados de forma que fique claro quais produtos se destacaram nas vendas de cada loja.

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs diretas para os arquivos CSV no GitHub (versão "raw")
urls = {
    "Loja 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv",
    "Loja 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv",
    "Loja 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv",
    "Loja 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"
}

resultados = []

# Processa cada loja
for loja, url in urls.items():
    df = pd.read_csv(url)

    # Conta vendas por produto
    contagem = df["Produto"].value_counts()

    # Produto mais vendido
    mais_vendido = contagem.idxmax()
    qtd_mais = contagem.max()

    # Produto menos vendido
    menos_vendido = contagem.idxmin()
    qtd_menos = contagem.min()

    resultados.append({"Loja": loja, "Tipo": "Mais Vendido", "Produto": mais_vendido, "Quantidade": qtd_mais})
    resultados.append({"Loja": loja, "Tipo": "Menos Vendido", "Produto": menos_vendido, "Quantidade": qtd_menos})

# Converte para DataFrame
df_resultados = pd.DataFrame(resultados)

# Exibe tabela de resultados
print(df_resultados)

# --- Gráfico somente com Matplotlib ---

###############################################################################################

# --- Gráfico horizontal ---
fig, ax = plt.subplots(figsize=(10, 6))

# Cores: verde (mais vendido), vermelho (menos vendido)
cores = {"Mais Vendido": "seagreen", "Menos Vendido": "firebrick"}

# Posições para as barras
y_pos = range(len(df_resultados))

# Barras horizontais
ax.barh(
    y_pos,
    df_resultados["Quantidade"],
    color=[cores[t] for t in df_resultados["Tipo"]]
)

# Rótulos nas barras
for i, (qtd, prod) in enumerate(zip(df_resultados["Quantidade"], df_resultados["Produto"])):
    ax.text(qtd + 0.5, i, f"{prod} ({qtd})", va='center', fontsize=9)

# Configurações de eixo e título
ax.set_yticks(y_pos)
ax.set_yticklabels(df_resultados["Loja"] + " - " + df_resultados["Tipo"])
ax.invert_yaxis()  # "Mais vendido" no topo do par
ax.set_xlabel("Quantidade Vendida")
ax.set_title("Produtos Mais e Menos Vendidos por Loja (Horizontal)", fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
```

##5.1. Atividade

Neste passo, vamos calcular o custo médio de frete para cada loja. O objetivo é entender quanto, em média, está sendo gasto com frete para cada uma das lojas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos arquivos CSV no GitHub
urls = {
    "Loja 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv",
    "Loja 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv",
    "Loja 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv",
    "Loja 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"
}

# Lista para armazenar resultados
medias_frete = []

# Processa cada loja
for loja, url in urls.items():
    df = pd.read_csv(url)

    # Calcula custo médio de frete
    media = df["Frete"].mean()
    medias_frete.append({"Loja": loja, "Custo Médio de Frete": round(media, 2)})

# Converte resultados para DataFrame
df_medias = pd.DataFrame(medias_frete)

# --- Exibe a tabela ---
print(df_medias)

# --- Visualização ---
plt.figure(figsize=(8,5))
plt.bar(df_medias["Loja"], df_medias["Custo Médio de Frete"], color="royalblue")

# Rótulos
for i, v in enumerate(df_medias["Custo Médio de Frete"]):
    plt.text(i, v + 0.2, f"R$ {v}", ha="center", fontsize=10)

plt.title("Custo Médio de Frete por Loja", fontsize=14, fontweight="bold")
plt.ylabel("Custo Médio (R$)")
plt.xlabel("Lojas")
plt.tight_layout()
plt.show()
```
